```python

from PIL import Image
import numpy as np
from skimage.filters import threshold_otsu
import skimage.exposure as imexp
import matplotlib
from skimage.color import rgb2gray
from matplotlib import pyplot as plt
from skimage.morphology import opening, closing, white_tophat,black_tophat
from skimage.morphology import disk
```

```python

#loading images
im = Image.open('samples/29  I.bmp')
im2 = Image.open('samples/30  .bmp')
im3 = Image.open('samples/34 .bmp')

npimg = np.array(im);
npimg2 = np.array(im2);
npimg3 = np.array(im3);
print(npimg.shape,npimg2.shape,npimg3.shape);


grayimg = rgb2gray(npimg)
grayimg2 = rgb2gray(npimg2)
grayimg3 = rgb2gray(npimg3)

testImg = grayimg2

selem = disk(6)
b_tophat = black_tophat(testImg, selem)
w_tophat = white_tophat(testImg, selem)
 

fig, axes= plt.subplots(1, 3, figsize=(30, 10), sharex=True, sharey=True)

axes[0].imshow(testImg,cmap='gray')
axes[0].set_title('original')
axes[0].axis('off')

axes[1].imshow(b_tophat,cmap='gray')
axes[1].set_title('black tophat')
axes[1].axis('off')

axes[2].imshow(w_tophat,cmap='gray')
axes[2].set_title('white tophat')
axes[2].axis('off')



fig2, axes2= plt.subplots(1, 3, figsize=(30, 10), sharex=True, sharey=True)

axes2[0].imshow(testImg,cmap='gray')
axes2[0].set_title('original')
axes2[0].axis('off')

axes2[1].imshow(testImg-b_tophat,cmap='gray')
axes2[1].set_title('black tophat')
axes2[1].axis('off')

axes2[2].imshow(testImg-w_tophat,cmap='gray')
axes2[2].set_title('white tophat')
axes2[2].axis('off')

plt.show()
```
